## Any economic crisis in the Eurozone manifests itself as social discontent over unemployment.

```python
unempeu = getfred(m4unempeu)
```

```python
plotfred(unempeu)
```

###### If we looked at youth unemployment, the situation looks dramatically worse. Likewise for single country unemployment, for example, from Greece or Portugal.

```python
#  What is the long-term unemployment trend?
unemptrend = regresstime(unempeu)
plotfred(unemptrend)
```

```python
#  Deviation from long-term trend:
unempdev = unempeu - unemptrend
plotfred(unempdev)
```

# Inflation in the Eurozone

```python
infleu = getfred(m4infleu)
```

```python
#  Compute inflation rate from levels: 
infreu = pcent(infleu, 12)
plotfred(infreu)
```

```python
stats(infreu)
```

```python
#  Inflation following Great Recession:
georet(infleu[t06], 12)
```

### The inflation rate plot looks a bit wild, but after the Great Recession we are still seeing inflation around long-term average of 2%. Now recent declines have set off deflation alarms, so here's the forecast which says inflation will not go sub-zero within one year:

```python
#  HW forecast on inflation level is more suitable than on inflation rate due to its trend properties:
holtfred(infleu, 12)
```

Using level data, inflation rate is forecasted to be 0.28% over the next 12 months.

# GDP numbers

```python
#  Real Eurozone GDP, interpolated monthly:
gdpeur = getfred(m4gdpeur)
plotfred(gdpeur)
```

```python
#  Real GDP growth after 2007 looks horrible:
georet(gdpeur['2007':], 12)
```

### Repricing Eurozone GDP in dollars for global perspective

```python
gdpeurusd = getfred(m4eurusd) * gdpeur
```

```python
plotfred(gdpeurusd)
```

```python
#  Real Eurozone GDP growth after 2007 in dollars:
georet(gdpeurusd['2007':], 12)
```

Notice the volatility introduced by FX risk, which in turn reduces the geometric return.

###### Now the comparison to the US GDP:

```python
gdpusr = getfred(m4gdpusr)
#  GDP data for US goes back far, so for a comparative look:
plotfred(gdpusr['1996':])
```

```python
#  Real US GDP growth after 2007 in dollars:
georet(gdpusr['2007':], 12)
```

###### Size of Eurozone / US economies using real GDP:

```python
eu2us = gdpeurusd / gdpusr
plotfred(eu2us)
```

### After 2007, Eurozone grew only 0.25% in dollar terms, whereas the US grew at 0.93% -- therefore, one can see the relative size of the Eurozone economy shrinking. Related fact: the allocation of Eurozone equities is also shrinking in global portfolios.

```python
plotfred(regresstime(eu2us))
```
