We shall follow the notation in Rob Hyndman, 2008, _Forecasting with Exponential Smoothing_, ignoring the seasonal aspect of the H-W model.

```python
Image(filename='holt-winters-equations.png', embed=True)
```

```python
#     GET DATA HERE ===============
wage = getfred( m4wage  )
```

```python
#     wage is in dollar/hour
inc = wage * 2000
#     assume 40 hours/week at 50 weeks/year 
#     to derive annual income for average nonfarm worker.
inc.describe()
```

```python
#  We also want to consider dollars in today's real terms:
deflator = getfred(m4defl)
```

```python
tail( deflator, 12)
```

```python
rinc = inc * deflator
```

```python
#  plot of inc is an uninteresting upward slope, 
#     and it is difficult to gauge worth of 
#     old dollars from decades ago.
plotdf(rinc, 'Rinc')
```

It is interesting that **real annual income ranges from 36K to 45K** in a non-monotonic way historically. Current annual income is mid-range. Real earnings can go down dramatically when inflation rages.

```python
#     We use T-bills + 10 bp as proxy for "risk-free" interest rate.
#                      10 bp barely takes the edge off near zero rates
#                      as divisors later :(
irate = (0.10 + monthly(getfred(d4bills))) / 100
irate.describe()
```

We see that irate never gets to zero or into negative territory, so calcuations will be fine numerically for division. Note how that 10 bp padding makes a vast difference at the zero end of the interest rate range: the difference in capital equivalence, say, between excessive \$35 million and *absurdly high \$350 million.*

```python
#  How much capital is required for interest income to equal annual wages?
rcap = rinc / (irate * 1000000)    
# Real capital expressed in $million
rcap.describe()
```

### Before 2008, the real capital equivalent stayed under \$1.5 million. However, with the Fed's zero rate policy, capital equivalent has zoomed past \$30 million! 

```python
#     Retain expensive result
holtdf = holt( rcap, alpha=0.2, beta=0.3 )
```

```python
start = '1960'

stats( holtdf[start:] )
#  Summary stats from custom start point:
```

```python
holtforecast(holtdf)
```

```python
plotholt(holtdf, 21)
```

```python
plotdf(holtdf['Y'][start:], 'H-W raw')
```

```python
plotdf(holtdf['Level'][start:], 'H-W level')
```

```python
plotdf(holtdf['Growth'][start:], 'H-W growth')
```
