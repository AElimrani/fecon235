```python
#  Some useful modules:       [ ] - print out EOF for reference.
from yi_1tools import *
from yi_fred import *
from yi_plot import *
from yi_timeseries import *
```

```python
#  Total US population in millions
pop = getfred(m4pop)/1000
```

```python
plotdf(pop)
```

```python
#  Fraction of population which works:
emppop = getfred(m4emppop)/100
```

Workers would be employed adults, which presumably exclude children (20% of pop) and most elderly persons (14% of pop). There is a dramatic drop in working% from about 64% in 2001 to about 59% recently.

```python
plotdf(emppop)
```

```python
#  Total US workers in millions:
workers = pop * emppop
```

```python
plotdf(workers)
```

Total population grows annually arount 1%, but the number of workers seems to be stabilizing around 185 million people.

```python
#  Deflator:
defl = getfred(m4defl)
```

```python
plotdf(defl)
```

```python
#  REAL GDP monthly in billions:
gdp = getfred(m4gdpus)
gdpr = defl * gdp
#  We do NOT use m4gdpusr directly because that is in 2009 dollars.
```

```python
plotdf(gdpr)
```

```python
#  REAL GDP per worker -- NOT per capita:
gdprworker = gdpr / workers
```

```python
plotdf(gdprworker)
#  plotted in thousands of dollars
```

```python
#  Annual income, not real, assuming 40 hours per week, 50 weeks per year:
inc = getfred(m4wage)*2000
#  But maybe working hours have decreased recently?
```

```python
#  REAL income in thousands per worker:
rinc = (defl * inc)/1000
```

```python
plotdf(rinc)
```

```python
#  Ratio of real GDP to real income per worker:
gdpinc = gdprworker / rinc
```

```python
plotdf(gdpinc)
```

### The ratio of real GDP to real income per worker has increased from 1.4 in the 1970s to 2.2 recently.

So each worker has become more productive (also with technological means) in producing national wealth. There is a noticeable temporary dip after the 2007 crisis.

## Appendix:  GDP implicit deflator vs. our aggregated inflation series

```python
#  GDP deflator is quarterly:
gdpdef = getfred('GDPDEF')
```

```python
plotdf(gdpdef)
```

```python
#  Our own deflator is based on m4infl, so:
qinfl = quarterly(getfred(m4infl))
```

```python
plotdf(qinfl)
```

```python
stat2(gdpdef['Y'], qinfl['Y'])
```

### The fit between the GDP deflator and our aggregated deflator is practically perfect.

Real GDP is currently based on 2009 dollars, but with our own deflator we can use plain GDP in current dollars and compute real GDP also in current dollars.
