#  ::  EURUSD

```python
eurusd = getfred(d4eurusd)
#  Note that the synthetic preprend back to 1971 is only for monthly data.
```

```python
plotfred(eurusd, 'd4 EURUSD')
```

### The 2006 support for EURUSD is currently at around 1.24. Its breach requires us to know the resistance from the synthetic euro series. 

```python
eursyn = getfred(m4eurusd)
```

```python
plotfred(eursyn, 'm4 EURUSD')
```

### The support dated 1985 implies the euro could easily descend to the 1.06 level based on the synthetic series composed of DEM and FRF.

```python
syntrend = trend(eursyn)
```

```python
plotfred(syntrend)
```

```python
#  Detrend:
plotfred( eursyn - syntrend )
```

```python
tailvalue(syntrend)
```

### Looking at the detrended series, a deviation of -0.3 (i.e. -3000 pips) is plausible from the trend. Thus the statistically 1.3280 - 0.3 = 1.0280 is realistic.

## Given ECB implicitly wants the euro to weaken (LSAP not yet in effect, as of October 2014), we see the long-term support between 1.03 and 1.06.

### Monetary policy is primary reason for weakening euro, however, Greece is expected to face default conditions beginning the start of 2016 -- the euro crisis will be revisited.

```python
#  Forecast monthly for next 24 months:
holtfred(eursyn)
```

```python
#  Look at geometric return since 2006:
georet(eurusd[t06:])
```

### Interestingly the annual geometric return post-Great-Recession is positive at 0.62% (which does not seem so in the charts visually). The volatility for EURUSD going back to 2006 is 10%.

```python
georet(eurusd['2013':])
```

### After 2013, we compute the annual depreciation at around -2.9%, i.e. -365 pips per annum.

```python
1.2618 * -0.029
```

### Let's look at post-2014 for the near term georet...

```python
georet(eurusd['2014':])
```

### So the volatility is declining, down to 5% level... but the geometric return is dramatic: -12.5% pa or -1577 pips per annum.

```python
1.2618 * -0.125
```

## At that near-term rate of descent, long-term support in the 1.03 to 1.06 region can be reached in less than two years.
