# Housing Starts

```python
#  in thousands of units
hs = getfred(m4housing)
```

```python
plotfred(hs)
```

### What makes most sense over the long-term is the housing starts per capita.

```python
#  population in thousands
pop = getfred(m4pop)
```

```python
#  Factor 100.00 converts operation to float and percentage terms
hspop = (hs * 100.00) / pop
```

```python
plotfred(hspop)
```

### So at the peaks, 1% of the US population gets allocated new housing monthly. The lowest point shown is after the Great Recession at 0.2%.

```python
stats(hspop)
```

```python
#  Let's fit the time trend:
hspop_trend = regresstime(hspop)
plotfred(hspop_trend)
```

```python
#  Holt-Winters forecast:
holtfred(hspop, beta=0.10)
```

```python
#  Finally to discern short-term economic cycles, we can detrend and normalize:
plotfred( detrendnorm( hspop ) )
```

### Historically, hspop looks way below the long-term mean of 0.61% -- however, by detrending, it appears that in the current economic cycle, housing starts has recovered and is back to fairly average levels.

# Constructing a Home Price Index

### The correlation between Case-Shiller 20-city vs 10-city is practically 1. Thus a mash-up is warranted to get data extended back to 1987. Now Case-Shiller is not dollar denominated so we use the median sales prices 2000 to mid-2014 from the National Assoc. of Realtors, see function gethomepx for explicit details.

```python
homepx = getfred(m4homepx)
```

```python
#  Case-Shiller is seasonally adjusted
plotfred(homepx)
```

```python
georet(homepx, 12)
```

### As an asset class, residential homes are not a great investment.

```python
stats(homepx)
```

## Real home prices

```python
defl = getfred(m4defl)
```

```python
homepxr = homepx * defl
```

```python
plotfred(homepxr)
```

```python
georet(homepxr, 12)
```

### Remember that the above does not include annual property taxes which does eat up a large portion of real price appreciation.

#### Perhaps home prices are only increasing because new stock of housing has been declining over the long-term.

# Affordability for typical home buyer

For a sketch, we assume a fixed premium for 30-y mortgages over 10-y Treasuries. Then figure out the number of hours needed just to pay off the interest on a home.

```python
mortpremium = 1.50
```

```python
mortgage = getfred(m4bond10) + mortpremium
```

```python
#  Yearly interest to be paid off:
interest = homepx * (mortgage / 100.00)
```

```python
#  Wage is in dollars per hour
wage = getfred(m4wage)
```

```python
#  Working hours to pay off just the interest:
interesthours = interest / wage
```

```python
#  Mortgage interest to be paid as portion of annual income:
plotfred(interesthours / float(2000) )
```

If we assume 2000 hours worked per year (40 hours for 50 weeks), we can see that interest payment can potentially take up to 50% of total annual pre-tax income. Currently that figure is about 20% so housing should be affordable, but the population is uncertain about the risk on taking on debt. Prospects of deflation adds to the fear of such risk.
