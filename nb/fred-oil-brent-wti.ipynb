# Brent vs. WTI

```python
brent = getfred(d4brent)
wti   = getfred(d4wti)
```

```python
#  BoW: Brent over WTI spread:
bow = brent - wti
```

```python
plotfred(bow)
```

### The difference between Brent and WTI is non-trivial. BoW can represent over 20% of the underlying oil price!

```python
#  Let's look the long-term BoW trend:
plotfred(trend(bow))
```

```python
#  Since the slope uses daily data, we annualize it:
0.0014257 * 256
```

##### Over the long haul, we can anticipate BoW to increase $0.36 annually per barrel, though there will be very large deviations.

# Oil: weighted average

```python
wtbrent = 0.50
wtwti   = 1 - wtbrent
oil = (wtbrent * brent) + (wtwti * wti )
```

```python
plotfred(oil, 'Oil weighted')
```

##### The Great Recession plunge from \$140 to \$40 is extraordinary.

## In fact, at equal weights, we have created two synthetic series called d4oil and m4oil.

```python
stats(oil)
```

## oil at $9.96 on 1998-12-10 :  HISTORICAL LOW

```python
tmin = '1998-12-10'
```

```python
georet(oil[tmin:])
```

## The geometric return since tmin is around 14% with volatility of 33% -- both very high relative to traditional assets.

### The technical support uptrend on the chart looks to be around $50 per barrel. Next we look at regresstime: 

```python
oiltrend = trend(oil[tmin:])
plotfred(oiltrend)
```

##### Now the detrended series using percentage deviation since tmin:

```python
plotfred(detrendpc(oil[tmin:]))
```

##### Prices can easily go +/- 40% off their statistical trend (about 1.2 std). 

```python
tailvalue(oiltrend) * 0.60
```

## The barrel bottom is in the region of \$50 to \$68.
